# Generating the Data

```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
import numpy as np
```

```python
seed(1)
```

```python
data = 5 * randn(100) + 50
```

```python
print(f"data mean {mean(data):.3f} , data std {std(data):.3f}")
```

# Visual Normal Check

## Histogram Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data)
plt.show()
```

## QQ Plot

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(data,line="s")
plt.show()
```

```python
std_normal = randn(100)
fig,ax = plt.subplots()
ax.scatter(sorted(std_normal),sorted(sorted((data - np.mean(data))/np.std(data))))
ax1 = ax.twinx()
ax.plot(sorted(std_normal),sorted(std_normal),
        color="red")
plt.plot()
```
