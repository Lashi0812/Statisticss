```python
from collections import defaultdict
```

what is the probability
that the sum of the dice equals seven?

```python
d={(i,j):i+j for i in range(1,7) for j in range(1,7)}
d
```

```python
dinv = defaultdict(list)
for pair,sums in d.items():
    dinv[sums].append(pair)
```

```python
dinv
```

```python
pmf = {i:len(j) / 36 for i, j in dinv.items()}
```

```python
pmf
```

what is the probability
that half the product of three dice will exceed the their sum? 

```python
 d={(i,j,k):((i*j*k)/2>i+j+k) for i in range(1,7) for j in range(1,7) for k in range(1,7)}
```

```python
dinv = defaultdict(list)
for pair,sums in d.items():
    dinv[sums].append(pair)
```

```python
dinv
```

```python
pmf = {i:len(j) / 6.0**3 for i, j in dinv.items()}
```

```python
pmf
```
